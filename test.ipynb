```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# Dane początkowe - punkty i krawędzie
points = np.random.rand(20, 2)  # Generowanie 10 losowych punktów
edges = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 0)]  # Przykładowe krawędzie

# Funkcja do aktualizacji animacji
def update(frame):
    plt.clf()
    # Rysowanie punktów
    
    # plt.scatter(points[:, 0], points[:, 1], color='white')
    plt.scatter(points[:, 0], points[:, 1], color='blue')
    # Rysowanie krawędzi
    for edge in edges:
        plt.plot([points[edge[0], 0], points[edge[1], 0]],
                 [points[edge[0], 1], points[edge[1], 1]], color='red')
    # Aktualizacja punktów
    # n = len(points)
    # a_ind = np.random.choice(n - frame) + frame
    # points[frame],points[a_ind] = points[a_ind],points[frame]
    
    for e in range(len(edges)) :
    # for i in range(n) :
    #     a_ind, b_ind = np.random.choice(n),np.random.choice(n)
    #     print(a_ind,b_ind)
    #     points[a_ind],points[b_ind] = points[b_ind],points[a_ind]
    
# Utworzenie animacji
fig = plt.figure()
ani = FuncAnimation(fig, update, frames=len(points), interval=1000)
ani.save('animation.gif', writer='pillow', fps=10)
```
